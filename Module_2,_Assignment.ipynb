```python
def clean_and_count(numbers):
    # Standard way to drop duplicates: convert to set and back to list
    unique_vals = list(set(numbers))

    # Building the report dictionary
    report = {
        "original_list": numbers,
        "unique_values": unique_vals,
        "total_unique": len(unique_vals)
    }

    return report

# Testing it out
my_nums = [10, 20, 10, 30, 40, 30, 50]
result = clean_and_count(my_nums)

print("Process Summary:")
print(result)
```

```python
def get_set_stats(set1=None, set2=None):
    # Fallback to empty sets if the user leaves one out
    s1 = set1 if set1 is not None else set()
    s2 = set2 if set2 is not None else set()

    # Calculating the standard set operations
    union = s1 | s2            # Using the | operator for union
    intersect = s1 & s2        # Using the & operator for intersection
    diff = s1 - s2             # Using the - operator for difference

    print(f"Set 1: {s1}")
    print(f"Set 2: {s2}")
    print(f"Combined (Union): {union}")
    print(f"Common (Intersection): {intersect}")
    print(f"Unique to Set 1 (Difference): {diff}")

    return union, intersect, diff

# Example: only providing one set to test the defaults
get_set_stats(set1={1, 2, 3, 4}, set2={3, 4, 5, 6})
```

```python
# A tuple with a string, an int, and a float
profile = ("Bappy", 24, 88.5)

# Unpacking them into descriptive variables
name, age, score = profile

# Creating a second tuple to compare against
other_profile = ("Muhit", 24, 90.0)

# Comparing: Python compares tuples index by index
is_same = profile == other_profile
is_less = profile < other_profile

print(f"Unpacked Data: Name is {name}, Age is {age}, Score is {score}")
print(f"Are the tuples identical? {is_same}")
print(f"Is profile 'smaller' than other_profile? {is_less}")

# QUICK NOTES ON LISTS VS TUPLES:
# 1. Mutability: This is the big one. Lists are 'mutable' (you can change,
#    add, or delete items). Tuples are 'immutable'â€”once they are
#    created, you can't touch them.
# 2. Performance: Because tuples can't change, Python handles them
#    faster in memory.
# 3. Intent: We use lists for collections of the same type of thing
#    (like a shopping list). We use tuples for 'records' where the
#    position of the data matters (like an (x, y) coordinate).
```
