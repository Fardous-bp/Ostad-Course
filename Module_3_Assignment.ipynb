```python
class Car:
    def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year

    def display_info(self):
        print(f"Car Details: {self.year} {self.brand} {self.model}")

# Creating the object
my_car = Car("Toyota", "Camry", 2022)
my_car.display_info()
```

```python
# --- Part A: Multilevel Inheritance ---
class Animal:
    def eat(self):
        print("This animal is eating.")

class Mammal(Animal):
    def walk(self):
        print("This mammal is walking.")

class Dog(Mammal):
    def bark(self):
        print("The dog is barking: Woof!")

# Testing the Dog object
my_dog = Dog()
my_dog.eat()   # Inherited from Animal
my_dog.walk()  # Inherited from Mammal
my_dog.bark()  # Own method


# --- Part B: Calculator and Method Overriding ---
class Calculator:
    def add(self, a, b, c=0):
        # Handles 2 or 3 parameters using a default value for c
        return a + b + c

class AdvancedCalculator(Calculator):
    def add(self, *args):
        # Overrides the method to handle any number of arguments
        return sum(args)

# Demonstrating functionality
calc = Calculator()
print(f"Basic Add (2 params): {calc.add(5, 10)}")
print(f"Basic Add (3 params): {calc.add(5, 10, 15)}")

adv_calc = AdvancedCalculator()
print(f"Advanced Add (5 params): {adv_calc.add(1, 2, 3, 4, 5)}")
```

```python
class BankAccount:
    def __init__(self, initial_balance):
        self.__balance = initial_balance  # Private attribute

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: ${amount}. New balance: ${self.__balance}")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew: ${amount}. Remaining: ${self.__balance}")
        else:
            print("Insufficient funds or invalid amount.")

# Subclasses for Polymorphism
class SavingsAccount(BankAccount):
    def account_type(self):
        print("This is a Savings Account.")

class CurrentAccount(BankAccount):
    def account_type(self):
        print("This is a Current Account.")

# Demonstration
savings = SavingsAccount(1000)
current = CurrentAccount(500)

# Polymorphism in action
accounts = [savings, current]
for acc in accounts:
    acc.account_type()

# Testing encapsulation
savings.deposit(200)
savings.withdraw(150)
# print(savings.__balance) # This would throw an error if uncommented
```
